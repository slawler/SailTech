```python
#Import Libraries
#Download tar.gz from https://github.com/slawler/PegasusLogs/tree/master/signalk

% matplotlib notebook
import numpy as np
import pandas as pd
from datetime import datetime

sample_data = "/home/slawler/Desktop/log.txt"
df = pd.read_csv(sample_data, header = None, sep = '\t')
df = df.pivot(index = 0, columns=1, values=2)
df = df.set_index(pd.to_datetime(df.index,format = '%Y-%m-%d %H:%M:%S'))
```

```python
speeds = df[['speedThroughWater','speedOverGround']]

speeds = speeds.dropna(axis=0, how='all')
speeds.describe()
speeds.plot()
```
